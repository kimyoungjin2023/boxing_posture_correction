```python
import cv2
import numpy as np
import tensorflow as tf

# PoseNet 모델 로드
model = tf.keras.models.load_model("C:/Users/user/boxing.png")

# 이미지 로드 및 전처리
def preprocess_image(image):
    image = cv2.resize(image, (257, 257))
    image = image.astype(np.float32) / 255.0
    image = np.expand_dims(image, axis=0)  # 배치 차원 추가
    return image

# 키포인트 예측 함수
def predict_keypoints(image):
    preprocessed_image = preprocess_image(image)
    keypoints = model.predict(preprocessed_image)
    return keypoints

# 이미지에서 키포인트 시각화 함수
def visualize_keypoints(image, keypoints):
    for point in keypoints[0]:
        x, y = int(point[0] * image.shape[1]), int(point[1] * image.shape[0])
        cv2.circle(image, (x, y), 5, (0, 255, 0), -1)
    return image

# 메인 함수
def main():
    image = cv2.imread("C:/Users/user/boxing.png")
    keypoints = predict_keypoints(image)
    output_image = visualize_keypoints(image, keypoints)
    
    cv2.imshow('Keypoints', output_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
